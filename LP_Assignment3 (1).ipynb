```python
df.loc[df['Chance of Admit ']>=0.8 , 'Chance of Admit '] = 1
df.loc[df['Chance of Admit ']<0.8 , 'Chance of Admit '] = 0
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
X = df.drop(columns = ['Chance of Admit '],axis = 1)
y = df['Chance of Admit ']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
clf = DecisionTreeClassifier().fit(X_train,y_train)
```

```python
preds = clf.predict(X_test)
preds
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, preds)*100
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, preds, labels=[0,1]))
```

```python
plt.figure(figsize=(14, 8))
tree.plot_tree(clf, filled=True, fontsize=16)
```

```python
clf = tree.DecisionTreeClassifier(criterion='gini', max_leaf_nodes=5).fit(X_train, y_train)
plt.figure(figsize=(14, 8))
tree.plot_tree(clf, filled=True, fontsize=16)
```
