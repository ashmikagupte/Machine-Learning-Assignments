```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = "/content/drive/MyDrive/company_sales_data .csv"
df = pd.read_csv(path)
```

```python
df
```

```python
df.info()
```

**1.Read Total profit of all months and show it using a line plot**

```python
df['total_profit']
```

```python
plt.plot(df['month_number'],df['total_profit'])
plt.xticks(np.arange(1, 13, 1))
plt.title("Company Profit per Month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
```

**2.Get total profit sales of all months and show line plot with the following Style properties**

```python
plt.figure(figsize=(8, 5))
plt.plot(df['month_number'],df['total_profit'],'--',marker='o', color='red', markerfacecolor='black', linewidth=3)
plt.xticks(np.arange(1, 13, 1))
plt.title("Company Sales Data of last year")
plt.xlabel("Month Number")
plt.ylabel("Profit in dollars")
plt.legend(['Profit data of last year'], loc='lower right')
```

**3.Read toothpaste sales data of each month and show it using a scatter plot**

```python
plt.scatter(df['month_number'],df['toothpaste'])
plt.xticks(np.arange(1, 13, 1))
plt.grid(linestyle = '--')
plt.title("Toothpaste Sales Data")
plt.ylabel("Number of units sold")
plt.xlabel("Month number")
plt.legend(['Toothpaste Sales Data'])
```

**4.Read face cream and facewash product sales data and show it using the bar chart**

```python
plt.figure(figsize=(12, 8))
df.plot(x="month_number", y=["facecream", "facewash"], kind="bar")
plt.xticks(np.arange(0, 13, 1))
plt.grid(linestyle = '--')
plt.title("Facewash and Facecream Sales Data")
plt.ylabel("Sales units in numbers")
plt.xlabel("Month number")
plt.legend(['Face Cream Sales Data','Face Wash Sales Data'], loc='upper left')
```

**5.Read sales data of bathing soap of all months and show it using a bar chart.**

```python
plt.bar(df['month_number'],df['bathingsoap'])
plt.xticks(np.arange(1, 13, 1))
plt.grid(linestyle='--')
plt.title("Bathing Soap Sales Data")
plt.ylabel("Sales units in number")
plt.xlabel("Month")
```

**6.Read the total profit of each month and show it using the histogram to see the most common profit ranges**

```python
plt.figure(figsize=(10, 5))
profitList = df['total_profit'].tolist()
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profitList, profit_range, label = 'Profit data')
plt.xlabel('Profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('Profit data')
```

**7.Read all product sales data and show it using the stack plot**

```python
plt.figure(figsize=(12, 8))
plt.stackplot(df['month_number'],df['facecream'],df['facewash'],df['toothpaste'],df['bathingsoap'],df['shampoo'],df['moisturizer'],colors =['m','c', 'r','k','g','y'])
plt.title("All Product Sales Data using Stack plot")
plt.xlabel("Month number")
plt.ylabel("Sales units in numbers")
plt.legend(['Face Cream','Face Wash','Toothpaste','Bathing Soap','Shampoo','Moisturizer'], loc='upper left')
```
